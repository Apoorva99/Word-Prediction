```python
import nltk
import pandas as pd
import numpy as np
import sys
import os
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import string
import re
```

```python
df = pd.read_csv('testdata.manual.2009.06.14.csv', header=None,parse_dates=[2])
print(df.head)
```

```python
df_cols=['Tweetpolarity', 'Tweetid','Tweetdate','query','user','TweetText']
df.columns = df_cols
df.columns
```

```python
df.drop(['Tweetpolarity', 'Tweetid','Tweetdate','query','user'], axis=1, inplace=True)
```

```python
#Removing @ $ 
for i in df.index:
     for i in df.index:
            sub = re.sub(r'@[A-Za-z0-9_]+','',df.TweetText[i])
            df["TweetText"][i]=sub
  
```

```python
df.head()
```

```python
#Removing '#' and other symbols
for i in df.index:
    sub = re.sub("[^a-zA-Z]", " ", df.TweetText[i])
    df["TweetText"][i]=sub  
```

```python
df.head()
```

```python
#Removing links from tweettexts
#Searche for a string which begins with http, followed by no or one 's', a colon and two slashes
#https://community.sophos.com/kb/en-us/117316
for i in df.index:
    sub = re.sub('https?://[A-Za-z0-9./]+','',df.TweetText[i])
    df["TweetText"][i]=sub
```
